```python
df_rain_types = df_rain.dtypes.to_frame('dtypes').reset_index()
df_rain_types
```

```python
df_rain_nonNull = df_rain.notnull().sum().to_frame('notnull').reset_index()
df_rain_nonNull
```

```python
df_rain_isnull = df_rain.isnull().sum().to_frame('isnull').reset_index()
df_rain_isnull
```

```python
df_rain_nunique = df_rain.nunique().to_frame('nunique').reset_index()
df_rain_nunique
```

```python
df2=pd.merge(df_rain_isnull,df_rain_nunique, on ='index')
df1=pd.merge(df_rain_types,df_rain_nonNull, on ='index')

df_rain_all_info = pd.merge(df1, df2)
df_rain_all_info.rename(columns={'index':"Columns"}, inplace=True)
df_rain_all_info
```

```python
df_rain.columns = df_rain.columns.str.lower()
df_rain
```

```python
df_rain.raintoday.replace('no', 'noRain', inplace=True)
df_rain.raintomorrow.replace('no', 'noRain', inplace=True)

df_rain
```

```python
df_rain.date = pd.to_datetime(df_rain.date)
df_rain.date = df_rain.date.dt.strftime('%d/%m/%Y')
df_rain.info()
```

```python
df_rain
```

```python
df_rain['difference'] = df_rain['maxtemp'] - df_rain['mintemp']
df_rain
```

```python
df_rain.drop('evaporation', axis=1, inplace=True)
df_rain.drop('sunshine', axis=1, inplace=True)
df_rain
```

```python
df_rain_most_rainy_day = df_rain.groupby(df_rain.location)['rainfall'].idxmax().to_frame('mostrainydays')
a = df_rain_most_rainy_day.mostrainydays.to_list()
df_rain.date[a]
```

```python
df_rain_Albury = df_rain[df_rain.location == 'Albury'].sort_values(by='maxtemp', ascending=False)
df_rain_Albury
```

```python
df_rain.isna().sum()
```

```python
df_rain.dropna(subset=['windgustdir'],inplace=True)
df_rain.reset_index(inplace=True)
df_rain
```

```python
df_rain2 = df.copy()
df_min = df_rain2.groupby('Location')[['Evaporation', 'Sunshine']].min()
df_min
```

```python
df_max = df_rain2.groupby('Location')[['Evaporation', 'Sunshine']].max()
df_max
```

```python
df_median = df_rain2.groupby('Location')[['Evaporation', 'Sunshine']].median()
df_median
```

```python
df_tri = pd.merge(df_min, df_max, on='Location', suffixes=('_min', '_max'))
df_location_evaporation_sunshine = pd.merge(df_tri, df_median, on='Location', suffixes=('', '_median'))
df_location_evaporation_sunshine
```
